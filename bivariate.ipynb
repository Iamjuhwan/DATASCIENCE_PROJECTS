```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
fuel = pd.read_csv("fuel-econ.csv")
fuel.head(10)
```

```python
print(fuel.shape)
```

```python
# using matplot lib
plt.scatter(data = fuel, x = 'displ', y = 'comb' );
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
# using seaborn
sb.regplot(data = fuel, x = 'displ', y = 'comb' );
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
# using seaborn
sb.regplot(data = fuel, x = 'displ', y = 'comb', fit_reg=False ); #fit_reg to turn off the regression line
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

### heatmap
it is used when there is two discrete variable
- good for discrete variable vs discrete variable
- good alternative to transparency for a lot of data
- bins size are important

```python

plt.hist2d(data = fuel, x = 'displ', y = 'comb' );
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
plt.hist2d(data = fuel, x = 'displ', y = 'comb' );
plt.colorbar()
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin=0.5 );
plt.colorbar()
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
plt.hist2d(data = fuel, x = 'displ', y = 'comb', cmin=0.5, cmap ='viridis_r' );
plt.colorbar()
plt.xlabel("Displacement (l)")
plt.ylabel("Combined fuel eff.");
```

```python
sedan = ['Minicompact Cars', 'Large Cars', 'Compact Cars', 'Subcompact Cars', 'Midsize Cars']
vclasses = pd.api.types.CategoricalDtype(ordered= True, categories=sedan)
fuel['VClass'] = fuel['VClass'].astype(vclasses);
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data=fuel, x= 'VClass', y= 'comb', color = base_color);

```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data=fuel, x= 'VClass', y= 'comb', color = base_color, inner= None);

plt.xticks(rotation = 15);
```

```python
## boxplot

base_color = sb.color_palette()[0]
sb.boxplot(data=fuel, x= 'VClass', y= 'comb', color = base_color);

plt.xticks(rotation = 15);
```

```python
fuel['trans_type'] = fuel['trans'].apply(lambda x: x.split()[0])
```

```python
#ct_counts =fuel.groupby(['VClass', 'trans_type']).size()

#ct_counts.reset_index(name = 'count')
#ct_counts['count'] = pd.to_numeric(ct_counts['count'], errors='coerce')  # Handle errors appropriately


ct_counts = fuel.groupby(['VClass', 'trans_type']).size()
ct_counts = ct_counts.reset_index(name = 'count')
#pivot_table = ct_counts.pivot(index='VClass', columns='trans_type', values='count')

ct_counts = ct_counts.pivot(index = 'VClass', columns = 'trans_type', values = 'count')
```

```python
sb.heatmap(ct_counts);
```
