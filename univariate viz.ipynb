```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
print(pokemon.shape)
```

```python
sb.countplot(data = pokemon, x = "generation_id");
```

```python
base_colour = sb.color_palette()[0]
```

```python
sb.countplot(data = pokemon, x = "generation_id", color = base_colour);
```

```python
sb.countplot(data = pokemon, x = "generation_id", color = base_colour, order = [5,1,3,4,2,7,6]);
```

```python
sb.countplot(data = pokemon, x = "type_1", color = base_colour);
plt.xticks(rotation = 90);
```

```python
type_order = pokemon['type_1'].value_counts().index
```

```python
sb.countplot(data = pokemon, y = "type_1", color = base_colour);
```

```python
sb.countplot(data = pokemon, y = "type_1", color =base_colour, order=type_order);
```
